```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#from the above data we analyse that the count of heart attack rate in males is greater
```

```python
age_bins = [0, 30, 40, 50, 60, 70, 80, 100]
age_labels = ['<30', '30-39', '40-49', '50-59', '60-69', '70-79', '80+']
df['Age_Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels)
plt.figure(figsize=(8, 5))
sns.barplot(x='Age_Group', y='Heart_Attack_Risk', data=df, estimator=sum, hue='Age_Group', palette='viridis', legend=False)
plt.title('Heart Attack Risk by Age Group')
plt.show()
```

```python
# Gender-wise Heart Attack Risk
plt.figure(figsize=(6, 4))
sns.countplot(x='Gender', hue='Heart_Attack_Risk', data=df, palette='Set1')
plt.title("Heart Attack Risk by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Risk Level", labels=["Low", "High"])
plt.show()
```

```python
# Visualizing the relationship between Diabetes and Heart Attack Risk
plt.figure(figsize=(10, 6))
sns.countplot(x='Heart_Attack_Risk', hue='Diabetes', data=df)
plt.title('Diabetes vs. Heart Attack Risk')
plt.xlabel('Heart Attack Risk')
plt.ylabel('Count')
plt.legend(title='Diabetes', loc='upper right', labels=['No', 'Yes'])
plt.grid(True)
plt.show()
```

```python
# Visualizing the relationship between Hypertension and Heart Attack Risk
plt.figure(figsize=(10, 6))
sns.countplot(x='Heart_Attack_Risk', hue='Hypertension', data=df)
plt.title('Hypertension vs. Heart Attack Risk')
plt.xlabel('Heart Attack Risk')
plt.ylabel('Count')
plt.legend(title='Hypertension', loc='upper right', labels=['No', 'Yes'])
plt.grid(True)
plt.show()
```

```python
# Visualizing the relationship between Physical Activity and Heart Attack Risk
plt.figure(figsize=(10, 6))
sns.countplot(x='Heart_Attack_Risk', hue='Physical_Activity', data=df)
plt.title('Physical Activity vs. Heart Attack Risk')
plt.xlabel('Heart Attack Risk')
plt.ylabel('Count')
plt.legend(title='Physical Activity', loc='upper right', labels=['No', 'Yes'])
plt.grid(True)
plt.show()
```
